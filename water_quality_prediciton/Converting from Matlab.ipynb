```python
import numpy as np
features = data['features'][0]
data_dict = {}
values = np.transpose(data['X_tr'][0][0])
for i,v in enumerate(features):
    data_dict[v[0]] = values[i]
    
df = pd.DataFrame(data_dict)

for x in range(1,len(data['X_tr'][0])):
    values = np.transpose(data['X_tr'][0][x])
    for i,v in enumerate(features):
        data_dict[v[0]] = values[i]
    temp_df = pd.DataFrame(data_dict)
    df = pd.concat([df,temp_df])
    
    
df.head()
```

```python
df.to_csv('water_x_train.csv',index = False)
```
